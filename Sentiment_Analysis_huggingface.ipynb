```python
from transformers import pipeline

# Load pre-trained model and tokenizer from Hugging Face
# The `pipeline` function automatically loads everything we need
classifier = pipeline('sentiment-analysis')

# Example reviews to analyze
reviews = [
    "I absolutely loved the movie, it was fantastic!",
    "The product quality was horrible, I want a refund.",
    "It was an okay experience, nothing special.",
    "The book was incredible, would highly recommend.",
    "This service is terrible and I am very disappointed."
]

# Perform sentiment analysis
results = classifier(reviews)

# Display the results
for review, result in zip(reviews, results):
    print(f"Review: {review}")
    print(f"Sentiment: {result['label']}, Confidence: {result['score']:.2f}")
    print()
```
